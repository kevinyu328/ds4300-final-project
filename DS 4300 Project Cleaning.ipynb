```python
import pandas as pd
import re
```

```python
columns = ['sentiment','id','date','flag','user','text']
```

```python
df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='latin-1',names = columns)
```

```python
df['sentiment'] = df['sentiment'].map({0: 'Negative', 4: 'Positive'})
df
```

```python
stop_words = []
```

```python
word_dict_negative = {}
word_dict_positive = {}

for text_row in range(len(df['text'])) :
    list_of_words = re.sub(r'[^\w\s]', '', df['text'][text_row]).split()
    list_of_words_2 = []
    for word in list_of_words :
        word = word.lower()
        word = word.replace('?','')
        word = word.replace('!','')
        word = word.replace('.','')
        word = word.replace('"','')
        word = word.replace("'",'')
        word = word.replace('#','')
        word = word.replace('$','')
        word = word.replace(':','')
        
        if '@' not in word and 'http' not in word :
            list_of_words_2.append(word)
            
    for word in list_of_words_2 :
        if df['sentiment'][text_row] == 'Negative' :
            word_dict_negative[word] = word_dict_negative.get(word,0) + 1
        else :
            word_dict_positive[word] = word_dict_positive.get(word,0) + 1
```

```python
df_word_probs = pd.DataFrame({'Negative':pd.Series(word_dict_negative),'Positive':pd.Series(word_dict_positive)})
df_word_probs = df_word_probs.reset_index()
df_word_probs = df_word_probs.rename(columns={"index": "word"})
df_word_probs = df_word_probs.fillna(0)
df_word_probs['Total'] = df_word_probs['Negative'] + df_word_probs['Positive']
#df_word_probs.to_csv('word_counts_10.csv',index=False)
df_word_probs_10 = df_word_probs[df_word_probs['Total'] > 9]
df_word_probs_10
```

```python
pair_dict_negative = {}
pair_dict_positive = {}
bad_rows_counter = 0


for text_row in range(len(df['text'])) :
    list_of_words = re.sub(r'[^\w\s]', '', df['text'][text_row]).split()
    list_of_words_2 = []
    for word in list_of_words :
        word = word.lower()
        word = word.replace('?','')
        word = word.replace('!','')
        word = word.replace('.','')
        word = word.replace('"','')
        word = word.replace("'",'')
        word = word.replace('#','')
        word = word.replace('$','')
        word = word.replace(':','')

        if '@' not in word :
            list_of_words_2.append(word)

    for word_ind in range(len(list_of_words_2) - 1) :
        word_1 = list_of_words_2[word_ind]
        word_2 = list_of_words_2[word_ind + 1]
        
        if df['sentiment'][text_row] == 'Negative' :
            pair_dict_negative[(word_1,word_2)] = pair_dict_negative.get((word_1,word_2),0) + 1
            
        else :
            pair_dict_positive[(word_1,word_2)] = pair_dict_positive.get((word_1,word_2),0) + 1
#     except :
#         bad_rows_counter +=1
#         continue
print(bad_rows_counter)
pair_dict_positive
```

```python
df_bigrams = pd.DataFrame({'Negative':pd.Series(pair_dict_negative),'Positive':pd.Series(pair_dict_positive)})
df_bigrams = df_bigrams.reset_index()
df_bigrams = df_bigrams.rename(columns={"level_0": "word 1","level_1": "word 2"})
df_bigrams = df_bigrams.fillna(0)
df_bigrams['Total'] = df_bigrams['Negative'] + df_bigrams['Positive'] 
df_bigrams
# df_bigrams.to_csv('bigram_counts.csv',index=False)
# df_bigramsb
```
